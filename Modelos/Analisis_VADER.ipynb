```python
df = pd.read_csv('/content/feedback_limpio.csv')
df[['Review_Text']].head(5)
```

```python
analyzer = SentimentIntensityAnalyzer()

def get_vader_sentiment(text):
    scores = analyzer.polarity_scores(text)
    compound = scores['compound']
    if compound >= 0.15:
        return 'positive'
    elif compound <= -0.05:
        return 'negative'
    else:
        return 'neutral'


df['Sentiment_Rating'] = pd.to_numeric(df['Rating'], errors='coerce').apply(lambda r: 'positive' if r >= 4 else 'neutral' if r == 3 else 'negative')
df['Sentiment_VADER'] = df['Review_Text'].apply(get_vader_sentiment)


print(classification_report(df['Sentiment_Rating'], df['Sentiment_VADER']))
print(confusion_matrix(df['Sentiment_Rating'], df['Sentiment_VADER']))
```

```python
df[['Review_Text', 'Rating', 'Sentiment_Rating', 'Sentiment_VADER']].to_csv('resultado_Vader_feedback.csv', index=False)
```
